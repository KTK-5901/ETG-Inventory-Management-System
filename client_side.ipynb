
## Read Inventory

```python
import json

fd = open("Inventory.json",'r')
r = fd.read()
fd.close()

Inventory = json.loads(r)
```

```python
Inventory
```

## Purchase

```python
import time

time_of_pur=str(time.ctime())

ui_ID = str(input("Enter Product ID: "))
if ui_ID in Inventory:
  ui_quantity= int(input("Enter Quantity: "))
  if ui_quantity <= Inventory[ui_ID]['Quantity']:
    print("Product: ", Inventory[ui_ID]['Name'])
    print("Price: ", Inventory[ui_ID]['Price'])
    print("Billing Amount: ", Inventory[ui_ID]['Price'] * ui_quantity)
    Inventory[ui_ID]['Quantity'] = Inventory[ui_ID]['Quantity'] - ui_quantity
  else:
    print("Not enough quantity")
else:
  print("ID doesn't exist")
```

###Inventory File

```python
js = json.dumps(Inventory)

fd = open("Inventory.json",'w')
fd.write(js)
fd.close()
```

###Sales file

```python
fd = open("Sales.json",'r')
txt = fd.read()
fd.close()
```

```python
Sales = json.loads(txt)
```

```python
s_no = len(Sales.keys()) + 1
Sales[s_no] = {"ID" : ui_ID, "Quantity" : ui_quantity, "Time" : time_of_pur,"Amount": Inventory[ui_ID]['Price'] * ui_quantity}

js=json.dumps(Sales)

fd = open("Sales.json",'w')
fd.write(js)
fd.close()
```
