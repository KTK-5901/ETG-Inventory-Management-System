####Add a new Item

```python
ID=str(input("Enter Product ID: "))
if ID in Inventory:
  print('This ID already exists')
else: 
  Name=str(input("Enter name: "))
  Price=int(input("Enter price: "))
  Quantity=int(input("Enter quantity: "))

Inventory[ID] = {"Name": Name, "Price":Price, "Quantity": Quantity}

js=json.dumps(Inventory)

fd = open("Inventory.json",'w')
fd.write(js)
fd.close()
```

####Replacing item

```python
fd = open("Inventory.json",'r')
txt = fd.read()
fd.close()

Inventory = json.loads(txt)

ID=str(input("Enter Product ID: "))
if ID in Inventory:
  print('Enter new information:-')
  Name=str(input("Enter name: "))
  Price=int(input("Enter price: "))
  Quantity=int(input("Enter quantity: "))
else: 
  print('This ID already exists')

Inventory[ID] = {"Name": Name, "Price":Price, "Quantity": Quantity}

js=json.dumps(Inventory)

fd = open("Inventory.json",'w')
fd.write(js)
fd.close()
```
